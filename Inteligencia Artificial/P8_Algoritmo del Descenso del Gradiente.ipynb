# PRACTICA No. 8. Gradiente Descendiente en 2D

### En esta practica aplicamos el Gradiente Descendiente para encontrar el minimo de una funcion en 2D. Representamos la siguiente funcion objetivo: f(x) = 2 * x * x * np.cos(x) - 5 * x como una funcion contigua

# Paso 1: Importar librerias necesarias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
```

# Paso 2: Definimos la funcion f(x)

```python
def f(x):
    return 2 * x * x * np.cos(x) - 5 * x
```

# Paso 3: Graficamos

```python
figsize = (6,4)
dpi = 150

fig, ax = plt.subplots(1,1, figsize = figsize, dpi = dpi)

# Definimos el intervalo 

t1 = np.arange(-5.0, 5.0, 0.1)

# Mostramos la grafica

ax.set_xlim (-5,5)
ax.set_ylim(-45,45)
ax.plot(t1,f(t1))

# x = [-1.0]
# y = [f(x[0])]
# ax.plot(x,y,'ro')

# print("X = ", x)
# print("Y = ", y)
```

# Algoritmo del descenso del gradiente

# Paso 1: Tomar un punto al azar X0

```python
fig, ax = plt.subplots(1,1, figsize = figsize, dpi = dpi)

# Definimos el intervalo 

t1 = np.arange(-5.0, 5.0, 0.1)

# Mostramos la grafica

ax.set_xlim (-5,5)
ax.set_ylim(-45,45)
ax.plot(t1,f(t1))

x = [-1.0]
y = [f(x[0])]
ax.plot(x,y,'ro')

print("X = ", x)
print("Y = ", y)
```

## Paso 2: Calcular el valor de la pendiente (derivada) f'(X0)

```python
# La derivada de f (la pendeinte)

def df(x):
    return 4 * x * np.cos(x) - 2 * x * x * np.sin(x) - 5

pendiente = df(x[0])
print ("Pendiente: ", [pendiente])
```

## Paso 3: Camine en direccion opuesta a la pendiente: x1 = x0 - eta * f'(x0)

### NOTA. Aqui eta es la tasa de aprendizaje que mencionamis previamente

```python
eta = 0.05

# eta = 0.7

x.append(x[0] - eta * pendiente)
y.append(f(x[0] - eta * pendiente))

print("X = ", x)
print("Y = ", y)
```

## Paso 4: Graficamos el 2do punto

```python
fig, ax = plt.subplots(1,1, figsize = figsize, dpi = dpi)

# Mostramos la grafica

ax.set_xlim (-5,5)
ax.set_ylim(-45,45)
ax.plot(t1,f(t1))
ax.plot(x,y,'ro')
```

## Paso 4: Graficamos m√°s puntos

```python
#eta = 0.05

ax.plot(t1,f(t1))

for etai in np.arange(0.05, 0.7,0.01):
    x.append(x[0] - etai * pendiente)
    y.append(f(x[0] - etai * pendiente))
    ax.plot(x,y,'ro')

# Mostramos la grafica
fig, ax = plt.subplots(1,1, figsize=figsize, dpi=dpi)
```
