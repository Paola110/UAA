# Paso 1: Importar librerias necesarias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
```

# Paso 2: Definimos la funcion f(x)

```python
def f(x):
    return 2 * x * x * np.cos(x) - 5 * x
```

# Paso 3: Graficamos

```python
figsize = (6,4)
dpi = 150

fig, ax = plt.subplots(1,1, figsize = figsize, dpi = dpi)

# Definimos el intervalo 

t1 = np.arange(-5.0, 5.0, 0.1)

# Mostramos la grafica

ax.set_xlim (-5,5)
ax.set_ylim(-45,45)
ax.plot(t1,f(t1))

x = [-1.0]
y = [f(x[0])]
ax.plot(x,y,'ro')

print("X = ", x)
print("Y = ", y)
```

# Paso 4:Calcular el valor de la pendiente (derivada) f'(x0)

```python
def df(x):
    return 4 * x * np.cos(x) - 2 * x * x * np.sin(x) - 5

pendiente = df(x[0])
print ("Pendiente: ", [pendiente])
```
