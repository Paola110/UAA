```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# carga los datos
datos = pd.read_csv('C:/Users/monts/OneDrive/Escritorio/UAA/Inteligencia Artificial/Recursos/EjemploEstudiantes.csv', sep=';', decimal=',', index_col=0)

# entrena el modelo
kmeans = KMeans(n_clusters=3, max_iter=2000, random_state=0)
grupos = kmeans.fit_predict(datos)

# grafica los clusters
plt.bar(range(1, 4), kmeans.cluster_centers_[:, 0], color='blue')  # centro de gravedad del cluster1
plt.bar(range(1, 4), kmeans.cluster_centers_[:, 1], color='red')   # centro de gravedad del cluster2
plt.bar(range(1, 4), kmeans.cluster_centers_[:, 2], color='green') # centro de gravedad del cluster3
plt.show()

# grafica los tres clusters completos
colors = plt.cm.viridis(np.linspace(0, 1, 3))  # Genera tres colores del mapa de color 'viridis'
for i in range(3):
    plt.bar(range(1, 4), kmeans.cluster_centers_[:, i], color=colors[i], width=0.5, label=f'Cluster {i+1}')
plt.xticks(range(1, 4), ['Cluster 1', 'Cluster 2', 'Cluster 3'])
plt.legend()
plt.show()

# En grupos verifico a que cluster pertenece cada fila de la tabla de datos
ndatos = pd.DataFrame(datos)
ndatos['Grupo'] = grupos

# guardar el resultado
ndatos.to_csv("C:/Users/monts/OneDrive/Escritorio/UAA/Inteligencia Artificial/Recursos/NDatos.csv")

```
