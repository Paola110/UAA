```python
import tkinter as tk
from tkinter import *
from tkinter import messagebox
from tkinter import ttk 

# ---------------------------- GUI constraints ------------------------------- #
WHITE = "#ffffff"
BLACK = "#000000"
FONT = ("Arial", 12)

# -------------------------- Base de conocimiento de vehículos ---------------- #
# Reglas para razonamiento hacia delante
reglas_adelante = [
    # Reglas para determinar el tipo de vehículo basado en el número de ruedas
    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['num_ruedas'] < 4, 
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'tipoDeVehiculo': 'cycle'})},
    
    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['num_ruedas'] == 4 and hechos_vehiculos['motor'] == True,
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'tipoDeVehiculo': 'automobil'})},
    
    # Reglas para identificar vehículos de tipo bicicleta
    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['tipoDeVehiculo'] == 'cycle' and hechos_vehiculos['num_ruedas'] == 2 and hechos_vehiculos['motor'] == False,
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'vehiculo': 'Bicicleta'})},
    
    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['tipoDeVehiculo'] == 'cycle' and hechos_vehiculos['num_ruedas'] == 3 and hechos_vehiculos['motor'] == False,
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'vehiculo': 'Triciclo'})},
    
    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['tipoDeVehiculo'] == 'cycle' and hechos_vehiculos['num_ruedas'] == 2 and hechos_vehiculos['motor'] == True,
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'vehiculo': 'Motocicleta'})},
    
    # Reglas para identificar vehículos de tipo automobile
    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['tipoDeVehiculo'] == 'automobil' and hechos_vehiculos['tamanio'] == 'grande' and hechos_vehiculos['num_puertas'] == 1,
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'vehiculo': 'Formula 1'})},

    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['tipoDeVehiculo'] == 'automobil' and hechos_vehiculos['tamanio'] == 'pequenio' and hechos_vehiculos['num_puertas'] == 2,
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'vehiculo': 'Carro deportivo'})},
    
    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['tipoDeVehiculo'] == 'automobil' and hechos_vehiculos['tamanio'] == 'mediano' and hechos_vehiculos['num_puertas'] == 4,
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'vehiculo': 'Sedan'})},
    
    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['tipoDeVehiculo'] == 'automobil' and hechos_vehiculos['tamanio'] == 'mediano' and hechos_vehiculos['num_puertas'] == 3,
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'vehiculo': 'MiniVan'})},
    
    {'condiciones': lambda hechos_vehiculos: hechos_vehiculos['tipoDeVehiculo'] == 'automobil' and hechos_vehiculos['tamanio'] == 'grande' and hechos_vehiculos['num_puertas'] == 4,
    'accion': lambda hechos_vehiculos: hechos_vehiculos.update({'vehiculo': 'Vehiculo deportivo utilitario'})}
]

# Reglas para razonamiento hacia atrás
reglas_atras = {
    'Bicicleta': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('tipoDeVehiculo') == 'cycle' and hechos_vehiculos.get('num_ruedas') == 2 and hechos_vehiculos.get('motor') == False}
    ],
    'Triciclo': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('tipoDeVehiculo') == 'cycle' and hechos_vehiculos.get('num_ruedas') == 3 and hechos_vehiculos.get('motor') == False}
    ],
    'Motocicleta': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('tipoDeVehiculo') == 'cycle' and hechos_vehiculos.get('num_ruedas') == 2 and hechos_vehiculos.get('motor') == True}
    ],
    'Formula 1': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('tipoDeVehiculo') == 'automobil' and hechos_vehiculos.get('tamanio') == 'grande' and hechos_vehiculos.get('num_puertas') == 1}
    ],
    'Carro deportivo': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('tipoDeVehiculo') == 'automobil' and hechos_vehiculos.get('tamanio') == 'pequenio' and hechos_vehiculos.get('num_puertas') == 2}
    ],
    'Sedan': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('tipoDeVehiculo') == 'automobil' and hechos_vehiculos.get('tamanio') == 'mediano' and hechos_vehiculos.get('num_puertas') == 4}
    ],
    'MiniVan': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('tipoDeVehiculo') == 'automobil' and hechos_vehiculos.get('tamanio') == 'mediano' and hechos_vehiculos.get('num_puertas') == 3}
    ],
    'Vehiculo deportivo utilitario': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('tipoDeVehiculo') == 'automobil' and hechos_vehiculos.get('tamanio') == 'grande' and hechos_vehiculos.get('num_puertas') == 4}
    ],
    'cycle': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('num_ruedas') < 4}
    ],
    'automobil': [
        {'condiciones': lambda hechos_vehiculos: hechos_vehiculos.get('num_ruedas') == 4 and hechos_vehiculos.get('motor') == True}
    ]
}
# -------------------------- Base de conocimiento de Dinosaurios ---------------- #
# Reglas para razonamiento hacia delante
reglas_adelante_dinosaurios = [

    # Reglas para determinar el tipo de habitad basado en el tipo extremidad
    {'condiciones': lambda hechos: hechos['tipo_extremidad'] == 'pata', 
    'accion': lambda hechos: hechos.update({'habitad': 'terrestre'})},

    {'condiciones': lambda hechos: hechos['tipo_extremidad'] == 'aleta', 
    'accion': lambda hechos: hechos.update({'habitad': 'acuatico'})},

    {'condiciones': lambda hechos: hechos['tipo_extremidad'] == 'ala', 
    'accion': lambda hechos: hechos.update({'habitad': 'cielo'})},

    # Determinar el tipo de dinosaurio en base al tipo de habitad, alimentacion, tamanio y numero de extremidades
    {'condiciones': lambda hechos: hechos['habitad'] == 'terrestre' and hechos['tipo_alimentacion'] == 'carnivoro' and hechos['tamanio'] == 'grande'and hechos['num_extremidades'] == 2,
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Giganotosaurio'})},
    
    {'condiciones': lambda hechos: hechos['habitad'] == 'terrestre' and hechos['tipo_alimentacion'] == 'herbivoro' and hechos['tamanio'] == 'grande'and hechos['num_extremidades'] == 4,
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Braquiosaurio'})},

    {'condiciones': lambda hechos: hechos['habitad'] == 'terrestre' and hechos['tipo_alimentacion'] == 'herbivoro' and hechos['tamanio'] == 'mediano'and hechos['num_extremidades'] == 4,
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Triceratops'})},

    {'condiciones': lambda hechos: hechos['habitad'] == 'terrestre' and hechos['tipo_alimentacion'] == 'carnivoro' and hechos['tamanio'] == 'pequenio'and hechos['num_extremidades'] == 2,
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Compsognathus'})},

    {'condiciones': lambda hechos: hechos['habitad'] == 'acuatico' and hechos['tipo_alimentacion'] == 'carnivoro' and hechos['tamanio'] == 'grande'and hechos['num_extremidades'] == 4,
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Mososaurio'})},
    
    {'condiciones': lambda hechos: hechos['habitad'] == 'acuatico' and hechos['tipo_alimentacion'] == 'carnivoro' and hechos['tamanio'] == 'grande'and hechos['num_extremidades'] == 2,
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Basilosaurio'})},

    {'condiciones': lambda hechos: hechos['habitad'] == 'acuatico' and hechos['tipo_alimentacion'] == 'carnivoro' and hechos['tamanio'] == 'mediano'and hechos['num_extremidades'] == 3,
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Megalodon '})},

    {'condiciones': lambda hechos: hechos['habitad'] == 'acuatico' and hechos['tipo_alimentacion'] == 'herbivoro' and hechos['tamanio'] == 'pequenio'and hechos['num_extremidades'] == 3,
    'accion': lambda hechos: hechos.update({'dinosaurio': 'ictiosaurio'})},

    {'condiciones': lambda hechos: hechos['habitad'] == 'cielo' and hechos['tipo_alimentacion'] == 'carnivoro' and hechos['tamanio'] == 'grande',
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Quetzal'})},

    {'condiciones': lambda hechos: hechos['habitad'] == 'cielo' and hechos['tipo_alimentacion'] == 'carnivoro' and hechos['tamanio'] == 'mediano',
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Argentavis'})},

    {'condiciones': lambda hechos: hechos['habitad'] == 'cielo' and hechos['tipo_alimentacion'] == 'carnivoro' and hechos['tamanio'] == 'pequenio',
    'accion': lambda hechos: hechos.update({'dinosaurio': 'Pteranodon'})},
]

reglas_atras_dinosaurios = {
    'Giganotosaurio': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'terrestre' and hechos.get('tipo_alimentacion') == 'carnivoro' and hechos.get('tamanio') == 'grande' and hechos.get('num_extremidades') == 2}
    ],
    'Braquiosaurio': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'terrestre' and hechos.get('tipo_alimentacion') == 'herbivoro' and hechos.get('tamanio') == 'grande' and hechos.get('num_extremidades') == 4}
    ],
    'Triceratops': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'terrestre' and hechos.get('tipo_alimentacion') == 'herbivoro' and hechos.get('tamanio') == 'mediano' and hechos.get('num_extremidades') == 4}
    ],
    'Compsognathus': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'terrestre' and hechos.get('tipo_alimentacion') == 'carnivoro' and hechos.get('tamanio') == 'pequenio' and hechos.get('num_extremidades') == 2}
    ],
    'Mososaurio': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'acuatico' and hechos.get('tipo_alimentacion') == 'carnivoro' and hechos.get('tamanio') == 'grande' and hechos.get('num_extremidades') == 4}
    ],
    'Basilosaurio': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'acuatico' and hechos.get('tipo_alimentacion') == 'carnivoro' and hechos.get('tamanio') == 'grande' and hechos.get('num_extremidades') == 2}
    ],
    'Megalodon': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'acuatico' and hechos.get('tipo_alimentacion') == 'carnivoro' and hechos.get('tamanio') == 'mediano' and hechos.get('num_extremidades') == 3}
    ],
    'Ictiosaurio': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'acuatico' and hechos.get('tipo_alimentacion') == 'herbivoro' and hechos.get('tamanio') == 'pequenio' and hechos.get('num_extremidades') == 3}
    ],
    'Quetzal': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'cielo' and hechos.get('tipo_alimentacion') == 'carnivoro' and hechos.get('tamanio') == 'grande'}
    ],
    'Argentavis': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'cielo' and hechos.get('tipo_alimentacion') == 'carnivoro' and hechos.get('tamanio') == 'mediano'}
    ],
    'Pteranodon': [
        {'condiciones': lambda hechos: hechos.get('habitad') == 'cielo' and hechos.get('tipo_alimentacion') == 'carnivoro' and hechos.get('tamanio') == 'pequenio'}
    ]
}

hechos = {}
vehiculo_inferido = None  # Almacenar el vehículo inferido
dinosaurio_inferido = None  # Almacenar el dinosaurio inferido

def razonamiento_hacia_adelante(hechos, reglas):
    cambios = True
    while cambios:
        cambios = False
        for regla in reglas:
            if regla['condiciones'](hechos):
                hechos_antes = hechos.copy()
                regla['accion'](hechos)
                if hechos_antes != hechos:
                    cambios = True
    return hechos


def razonamiento_hacia_atras(hechos, objetivo):
    if 'vehiculo' in hechos and hechos['vehiculo'] == objetivo:
        return True
    if 'dinosaurio' in hechos and hechos['dinosaurio'] == objetivo:
        return True
    
    # Verificar las reglas de retroceso para el objetivo dado
    if objetivo in reglas_atras:
        for regla in reglas_atras[objetivo]:
            # Comprobar si se cumplen las condiciones de la regla
            if regla['condiciones'](hechos):
                # Inferir el objetivo si se cumplen todas las condiciones
                hechos[objetivo.lower()] = objetivo
                return True
    return False

# Función para mostrar preguntas una por una en ventanas flotantes

def hacer_pregunta(pregunta, opciones, clave_hecho, conversion_func=None):
    base_conocimiento = var_base.get()
    top = tk.Toplevel(ventana)
    top.title("Pregunta")

    top.configure(bg=WHITE)

    label_pregunta = tk.Label(top, text=pregunta, bg=WHITE, fg=BLACK, font=FONT)
    label_pregunta.pack(pady=10)

    # Variable para almacenar la respuesta seleccionada
    var_respuesta = tk.StringVar()

    def ajustar(combobox):
        ancho = max(len(opcion) for opcion in combobox['values']) + 3
        combobox.config(width=ancho)

    menu_opciones = ttk.Combobox(top, textvariable=var_respuesta, values=opciones, state="readonly")
    menu_opciones.set(opciones[0])
    menu_opciones.pack(pady=10)
    menu_opciones.configure(font=FONT)

    ajustar(menu_opciones)

    # Función de respuesta
    def responder():
        respuesta = var_respuesta.get()

        if conversion_func:
            respuesta = conversion_func(respuesta)
        hechos[clave_hecho] = respuesta
        top.destroy()

        if base_conocimiento == "Vehículos":
            if not any(key not in hechos for key in ['num_ruedas', 'motor', 'tamanio', 'num_puertas']):
                ejecutar_inferencia()
            else:
                procesar_preguntas_vehiculos()
        elif base_conocimiento == "Dinosaurios":
            if not any(key not in hechos for key in ['tipo_extremidad', 'tipo_alimentacion', 'tamanio', 'num_extremidades']):
                ejecutar_inferencia()
            else:
                procesar_preguntas_dinosaurios()

    boton_responder = tk.Button(top, text="Aceptar", command=responder, bg=WHITE, fg=BLACK, font=FONT)
    boton_responder.pack(pady=10)

    # Centrar la ventana emergente sobre la ventana principal
    top.update_idletasks()  # Actualiza para obtener el tamaño
    ancho = top.winfo_width()
    alto = top.winfo_height()
    x = (top.winfo_screenwidth() // 2) - (ancho // 2)
    y = (top.winfo_screenheight() // 2) - (alto // 2)
    top.geometry(f"{ancho}x{alto}+{x}+{y}")


# Función para procesar preguntas sobre vehículos
def procesar_preguntas_vehiculos():
    if 'num_ruedas' not in hechos:
        hacer_pregunta("¿Cuántas ruedas tiene el vehículo?", ["2", "3", "4"], 'num_ruedas', conversion_func=int)
    elif 'motor' not in hechos:
        hacer_pregunta("¿Tiene motor?", ["Sí", "No"], 'motor', conversion_func=lambda x: x == "Sí")
    elif hechos['num_ruedas'] == 4 and 'tamanio' not in hechos:
        hacer_pregunta("¿Qué tamaño tiene el vehículo?", ["pequeño", "mediano", "grande"], 'tamanio')
    elif hechos['num_ruedas'] == 4 and 'num_puertas' not in hechos:
        hacer_pregunta("¿Cuántas puertas tiene el vehículo?", ["1", "2", "3", "4"], 'num_puertas', conversion_func=int)
    else:
        ejecutar_inferencia()

# Función para procesar preguntas sobre dinosaurios
def procesar_preguntas_dinosaurios():
    if 'tipo_extremidad' not in hechos:
        hacer_pregunta("¿Qué tipo de extremidad tiene?", ["pata", "aleta", "ala"], 'tipo_extremidad')
    elif 'tipo_alimentacion' not in hechos:
        hacer_pregunta("¿Qué tipo de alimentación tiene?", ["carnivoro", "herbivoro"], 'tipo_alimentacion')
    elif 'tamanio' not in hechos:
        hacer_pregunta("¿Qué tamaño tiene?", ["pequeño", "mediano", "grande"], 'tamanio')
    elif 'num_extremidades' not in hechos:
        hacer_pregunta("¿Cuántas extremidades tiene?", ["2", "4", "3"], 'num_extremidades', conversion_func=int)
    else:
        ejecutar_inferencia()

# Ejecutar razonamiento basado en la selección del usuario
def ejecutar_inferencia():
    global vehiculo_inferido, dinosaurio_inferido
    base_conocimiento = var_base.get()
    tipo_razonamiento = var_tipo.get()

    # Definir reglas según la base de conocimiento seleccionada
    if base_conocimiento == "Vehículos":
        reglas_seleccionadas = reglas_adelante
    elif base_conocimiento == "Dinosaurios":
        reglas_seleccionadas = reglas_adelante_dinosaurios
    
    proceso = ""
    if tipo_razonamiento == "Hacia adelante":
        proceso = razonamiento_hacia_adelante(hechos, reglas_seleccionadas)
        if base_conocimiento == "Vehículos":
            vehiculo_inferido = hechos.get('vehiculo', None)
            resultado_texto.set(f"{vehiculo_inferido if vehiculo_inferido else 'No se pudo inferir'}")
        elif base_conocimiento == "Dinosaurios":
            dinosaurio_inferido = hechos.get('dinosaurio', None)
            resultado_texto.set(f"{dinosaurio_inferido if dinosaurio_inferido else 'No se pudo inferir'}")
    else:
        if base_conocimiento == "Vehículos":
            if vehiculo_inferido is None:
                # Si no hay resultado anterior, primero ejecutar hacia adelante
                proceso = razonamiento_hacia_adelante(hechos, reglas_seleccionadas)
                vehiculo_inferido = hechos.get('vehiculo', None)  # Almacenar el vehículo inferido
            
            if vehiculo_inferido:
                # Tomar el resultado del razonamiento hacia adelante como objetivo
                objetivo = vehiculo_inferido
                if razonamiento_hacia_atras(hechos, objetivo):
                    proceso += f"\nRazonamiento hacia atrás confirmado para: {objetivo}"
                else:
                    proceso += f"\nNo se pudo confirmar el resultado hacia atrás."
            else:
                proceso += "\nNo hay un vehículo inferido para verificar hacia atrás."
            
        elif base_conocimiento == "Dinosaurios":
            if dinosaurio_inferido is None:
                # Si no hay resultado anterior, primero ejecutar hacia adelante
                proceso = razonamiento_hacia_adelante(hechos, reglas_seleccionadas)
                dinosaurio_inferido = hechos.get('dinosaurio', None)  # Almacenar el vehículo inferido
            
            if dinosaurio_inferido:
                # Tomar el resultado del razonamiento hacia adelante como objetivo
                objetivo = dinosaurio_inferido
                if razonamiento_hacia_atras(hechos, objetivo):
                    proceso += f"\nRazonamiento hacia atrás confirmado para: {objetivo}"
                else:
                    proceso += f"\nNo se pudo confirmar el resultado hacia atrás."

    # Mostrar el proceso
    #proceso_texto.delete(1.0, tk.END)  # Limpiar el widget Text
    proceso_texto.insert(tk.END, f"Proceso:\n{proceso}")  # Insertar nuevo texto

def ejecutar():
    global vehiculo_inferido, dinosaurio_inferido
    if var_tipo.get() == "Hacia adelante":
        hechos.clear()  # Limpiar los hechos
        vehiculo_inferido = None
        dinosaurio_inferido = None
        proceso_texto.delete(1.0, tk.END)  # Elimina todo el texto en el widget Text
        
        # Comenzar las preguntas según la base seleccionada
        if var_base.get() == "Vehículos":
            procesar_preguntas_vehiculos()
        else:
            procesar_preguntas_dinosaurios()
    elif var_tipo.get() == "Hacia atrás":
        if vehiculo_inferido:
            ejecutar_inferencia()
        elif dinosaurio_inferido:
            ejecutar_inferencia()
# ---------------------------- UI SETUP ------------------------------- #

ventana = tk.Tk()
ventana.title("Agente inteligente")
ventana.geometry("600x500")
ventana.config(padx=20, pady=20, background=WHITE)

# Variables para las selecciones
var_base = tk.StringVar(value="Vehículos")
var_tipo = tk.StringVar(value="Hacia adelante")

# Opción de Base de Conocimientos
# Barra de menú
barra_menus = tk.Menu(ventana)

# Menú de Base de Conocimientos
menu_base = tk.Menu(barra_menus, tearoff=False)

# Opciones de Base de Conocimientos
opciones_base = ["Vehículos", "Dinosaurios"]  # Puedes añadir más bases de conocimiento

# Función para cambiar la selección de base de conocimientos
def seleccionar_base(opcion):
    var_base.set(opcion)
    print(f"Base seleccionada: {opcion}")

# Añadir las opciones al menú
for opcion in opciones_base:
    menu_base.add_command(label=opcion, command=lambda opt=opcion: seleccionar_base(opt))

# Añadir el menú de "Base de Conocimientos" a la barra de menús
barra_menus.add_cascade(menu=menu_base, label="Base de Conocimientos")

# Configurar la ventana para usar la barra de menús
ventana.config(menu=barra_menus)

# Opción de Tipo de Razonamiento
label_tipo = tk.Label(ventana, text="Selecciona el tipo de razonamiento:",  font=FONT, fg=BLACK, bg=WHITE)
label_tipo.grid(column=0, row=0, sticky="w", pady=(10, 5))

# Radio Buttons para Tipo de Razonamiento
radio_hacia_adelante = tk.Radiobutton(ventana, text="Hacia adelante", variable=var_tipo, value="Hacia adelante", font=FONT, fg=BLACK, bg=WHITE)
radio_hacia_adelante.grid(column=0, row=1, sticky="w")

radio_hacia_atras = tk.Radiobutton(ventana, text="Hacia atrás", variable=var_tipo, value="Hacia atrás",  font=FONT, fg=BLACK, bg=WHITE)
radio_hacia_atras.grid(column=0, row=2, sticky="w")


# Reset learning button
def resetLearningSelect():
    var_tipo.set(None)

resetButton = Button(text="Resetear opciones", command=resetLearningSelect,  font=FONT)
resetButton.grid(column=0, row=3, pady=(10,10), sticky="w")

# Reset output button
def resetOutput():
    proceso_texto.delete("1.0", "end")

resetButton = Button(text="Resetear salida", command=resetOutput,  font=FONT)
resetButton.grid(column=1, row=3, pady=(10, 10), sticky="w")

# Etiqueta para el resultado
label_resultado_titulo = tk.Label(ventana, text="Resultado:", font=FONT, fg=BLACK, bg=WHITE)
label_resultado_titulo.grid(column=1, row=0, sticky="w", padx=(10, 10))

# Campo de texto no editable para mostrar el resultado
resultado_texto = tk.StringVar()
entry_resultado = tk.Entry(ventana, textvariable=resultado_texto, font=FONT, fg=BLACK, bg=WHITE, state="readonly")
entry_resultado.grid(column=1, row=1, sticky="w", padx=(10, 10))

# Caja de texto para mostrar el proceso
proceso_texto = tk.Text(ventana, height=10, width=50,  bg = "light blue")
proceso_texto.grid(column=0, row=6, padx=20, pady=(30, 30), sticky="nsew", columnspan=2)

# Botón para iniciar la ejecución
boton_iniciar = tk.Button(ventana, text="Iniciar", command=ejecutar,  font=FONT)
boton_iniciar.grid(column=0, row=7, pady=(10,10), columnspan=2)

# Configuración adicional para los elementos
# Hacer que la fila 5 (donde está el Text) se expanda al redimensionar
ventana.grid_rowconfigure(6, weight=1)

# Asignar más peso a la columna 0 para que se expanda más que la columna 1
ventana.grid_columnconfigure(0, weight=3)  # Columna 0 se expande más
ventana.grid_columnconfigure(1, weight=3)  # Columna 1 se expande menos

ventana.update_idletasks()
ancho = ventana.winfo_width()
alto = ventana.winfo_height()
x = (ventana.winfo_screenwidth() // 2) - (ancho // 2)
y = (ventana.winfo_screenheight() // 2) - (alto // 2)
ventana.geometry(f"{ancho}x{alto}+{x}+{y}")


ventana.mainloop()
```
